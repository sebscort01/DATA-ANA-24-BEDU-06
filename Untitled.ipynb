```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FILENAME = 'tmp/Ejemplo_02_03_Cybersecurity_Dataset.csv'

# Cargar nuestro dataset
cybersecurity_data = pd.read_csv(FILENAME)
cybersecurity_data.head()

g = sns.FacetGrid(cybersecurity_data, col='attack_type', col_wrap=3, height=4)
g.map(sns.scatterplot, 'severity', 'response_time', color='b')

g.fig.suptitle('Severidad VS Tiempo de respuesta en ataques de ciber seguridad', fontsize=14)
g.fig.tight_layout(rect=[0, 0, 1, 0.95])


# mostrar grafico
plt.show()
```
